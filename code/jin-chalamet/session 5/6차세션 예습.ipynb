[경사하강법]
"어느 방향으로 어떻게 갈 것인가"를 결정

'어느 방향'
-미분 값이 결정

'어떻게'
-learning rate가 결정
-learning rate가 너무크면 모델은 발산하게 되고, learning rate가 너무 작다면 모델이 수렴하는 데 너무 오랜 시간이 걸리게 됨.

'경사하강법의 문제점'
-local minima 문제
손실 함수의 전역 최솟값이 아닌 지역 극솟값으로 수렴하는 문제
-적절한 크기의 step size를 찾지 못해 생기는 문제
학습률을 너무 크게 설정하면 다른 곳으로 발산하는 현상, 반대의 경우에는 시간이 오래 걸림.
-해결법; 모멘텀
기울기에 관성 부여, 작은 기울기는 쉽게 넘어갈 수 있도록 만듦!

[규제선형모델]
모델의 오버피팅을 막기 위해 등장한 모델

'규제선형모델의 종류'
1. 릿지회귀 : 계수를 제고한 기준으로 규제 적용
2. 라쏘회귀 : 계수의 절댓값 기준으로 규제를 적용
3. 엘라스틱넷 회귀 : L2 규제와 L1 규제를 결합한 회귀

'편향과 분산의 TRADE-OFF'
과소적합; 높은 편향/ 낮은 분산
과대적합; 낮은 편향/ 높은 분산

['SCALING']
데이터 전처리 과정 중 하나로, 모든 피쳐들의 데이터 분포나 범위를 동일하게 조정하는 과정

'스케일링종류'
1. StandardScaler : 평균을 0, 분산을 1로 맞추는 표준화. (선형모델이나 신경망과 같은 알고리즘에서 잘 작동, 회귀보다 분류에 유용)
2. MinMaxScaler : 모든 피쳐들이 0과 1사이의 데이터 값을 갖도록 변환. (분류보다 회귀에 유용. 데이터의 최소/최대값을 알 때 사용함)
3. MaxAbsScaler : 절대값의 최소값이 0, 절대값의 최대값이 1이 되도록 스케일링 (큰 이상치에 민감할 수 있음)
4. RobustScaler :  데이터의 중앙값이 0, IQE=1이 되도록 스케일링
5. Normalizer : 각 데이터 포인트의 크기를 1로 만드는 방법

[차원축소]
차원이 크면 시각화가 어렵고, 분석하기에도 어렵기 때문!

'PCA'
차원을 줄이기 위해 축을 회전시켜 새로운 축을 찾아냄.
데이터를 축에 사영했을 때 가장 높은 분산을 가지는 데이터의 축을 찾는 것!!

'LDA'
지도학습의 분류에서 사용하기 쉽도록 개별 클래스를 분별할 수 있는 기준을 최대한 유지하면서 차원축소

'SVD'
M*N 형태의 행렬을 분해

'NMF'
하나의 객체정보를 음수를 포함하지 않는 두 개의 부분 정보로 인수분해

'차원축소의 이유'
- 차원이 커지면서 두 데이터 간의 거리가 멀어지고 밀도가 급격히 줄어드는 현상을 막기 위해
-오버피팅의 문제를 피하기 위해 (복잡성 UP, 효율성 DOWN)
- 시각화의 용이를 위해
