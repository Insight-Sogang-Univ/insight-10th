## 과제 2

공분산과 상관계수에 대해 자세히 설명하시오

정답:

공분산(covariance)은 조사하고자 하는 두 변수가 어떠한 관계(양/음)를 가지는지와 그 크기를 표현하는데 사용된다. Cov(x,y)의 값이 0보다 크다면 양의 관계, 0보다 작다면 음의 관계를 가지는 것이다.

상관계수(correlation)는 두 변수 간의 선형적인 관계를 나타내는 값이며, -1과 1사이의 값을 가진다. 이는 공분산을 -1과 1 사이로 표준화한 값으로, 공분산과 달리 x와 y의 절대적인 단위의 크기에 영향을 받지 않는다.

## 과제3

확률분포의 예시 7가지를 쓰시오

정답:

(1) 정규분포 : 종 묘양, 좌우 대칭, 위치는 평균에 의해 정해지며 모양은 표준편차의 크기에 따라 결정됨

(2) T분포 : 0을 중심으로 종형의 모습을 가진 대칭형, 꼬리가 정규분포보다 두꺼우며 자유도에 따라 모습이 변한다.

(3) 카이제곱분포 : 정규분포의 제곱, 범주형 데이터 분석이나 카이제곱 검정과 같은 통계적 검정에 활용

(4) F분포 : 두 개의 카이제곱분포의 비율에 관련된 확률분포, 자유도가 증가할수록 확률밀도 함수의 모양이 정규분포에 가까워짐

(5) 이항분포 : 연속된 n번의 독립시행에서 각 시행이 성공할 확률 p를 가질때 만들어지는 이산 확률 분포

(6) 베르누이분포 : 이항분포에서 시행이 1번인 경우

(7) 포아송분포 : 단위 시간 안에 어떤 사건이 몇 번 일어날 것인지를 표현하는 이산 확률 분포

  - 귀무가설과 대립가설을 세운다.

2단계 : 유의 수준 설정.
  - 정해진 수치를 얼마나 벗어났을 때 귀무가설이 오류라고 인정할 것인지 판단하는 기준을 세움.
  - 이 값이 커질수록 대립가설이 채택될 확률이 높아진다.

3단계 : 검정통계량 산출.
  - 가설 검정의 결과를 판단하는데 사용
  - 데이터가 분포나 특징에 따라 z통계량, T통계량 등 어떤 검정통계량을 사용할지 정해야 함.

4단계 : 가설 기각 / 채택 판단
  - 검정통계량이 확률분포 어디에 위치하는지를 파악해 가설 기각 및 채택

  - 검정통계량에서 유도할 수 있는 유의확률을 계산한 후, 유의확률이 유의수준보다 클 시 귀무가설을 채택.

## 과제 6

베이즈 정리 공식에 대해 쓰시오

정답:

베이저 정리(Bayes' Theorem) : 조건부 확률을 이용하여 사전 정보가 주어졌을 때 그 정보를 통해 새로운 정보를 추론하는 방법 -> 사전 확률을 기반으로 사후 확률을 계산하는 형태

P(A|B) = P(B|A)P(A) / P(B)
