## EDA/ÏãúÍ∞ÅÌôî
### 'Default'ÏôÄÏùò Í¥ÄÍ≥Ñ ÎÉÖÎã§ Ï∞çÏñ¥Î≥¥Í∏∞
- Î≤îÏ£ºÌòï -> bar plot
- ÏàòÏπòÌòï -> histplot

ÏúºÎ°ú 'Default'ÏôÄÏùò Í¥ÄÍ≥ÑÎ•º Ï∞çÏñ¥Î≥¥ÏïòÏùå
- Î≤îÏ£ºÌòï: `'EmploymentType'`, `'MaritalStatus'`, `'HasMortgage'`, `'HasDependents'`, `'LoanPurpose'`, `'HasCoSigner'`
- ÏàòÏπòÌòï:`'Age'`, `'Income'`, `'LoanAmount'`, `'CreditScore'`, `'MonthsEmployed'`, `'NumCreditLines'`,
    `'LoanTerm'`, `'Education'`,`'InterestRate'`, `'DTIRatio'` 

```python
loan = pd.DataFrame(loan)

# 'EmploymentType'Ïóê Îî∞Î•∏ 'Default'Ïùò ÎπÑÏú®ÏùÑ Í≥ÑÏÇ∞
result = loan.groupby('EmploymentType')['Default'].mean().reset_index()

# 2x3 ÏÑúÎ∏åÌîåÎ°Ø ÏÉùÏÑ±
fig, axs = plt.subplots(2, 3, figsize=(15, 8))
fig.suptitle('Default Rate', fontsize=16)

# Ï≤´ Î≤àÏß∏ ÏÑúÎ∏åÌîåÎ°ØÏóê Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞ - Employment Type
axs[0, 0].bar(result['EmploymentType'], result['Default'], color='skyblue')
axs[0, 0].set_xlabel('Employment Type')
axs[0, 0].set_ylabel('Default Rate')
axs[0, 0].set_title('Employment Type - Default rate')
axs[0, 0].tick_params(axis='x', rotation=45)

# ÎÇòÎ®∏ÏßÄ ÏÑúÎ∏åÌîåÎ°ØÏóê ÌïÑÏöîÌïú Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
# axs[0, 1].bar(...) - MaritalStatus
result = loan.groupby('MaritalStatus')['Default'].mean().reset_index()

axs[0, 1].bar(result['MaritalStatus'], result['Default'], color='skyblue')
axs[0, 1].set_xlabel('Marital Status')
axs[0, 1].set_ylabel('Default Rate')
axs[0, 1].set_title('Marital Status - Default rate')
axs[0, 1].tick_params(axis='x', rotation=45)

# axs[0, 2].bar(...)
result = loan.groupby('HasMortgage')['Default'].mean().reset_index()

axs[0, 2].bar(result['HasMortgage'], result['Default'], color='skyblue')
axs[0, 2].set_xlabel('Has Mortgage')
axs[0, 2].set_ylabel('Default Rate')
axs[0, 2].set_title('HasMortgage - Default rate')
axs[0, 2].tick_params(axis='x', rotation=45)

# axs[1, 0].bar(...) - HasDependents
result = loan.groupby('HasDependents')['Default'].mean().reset_index()

axs[1, 0].bar(result['HasDependents'], result['Default'], color='skyblue')
axs[1, 0].set_xlabel('Has Dependents')
axs[1, 0].set_ylabel('Default Rate')
axs[1, 0].set_title('HasDependents - Default rate')
axs[1, 0].tick_params(axis='x', rotation=45)

# axs[1, 1].bar(...)
result = loan.groupby('LoanPurpose')['Default'].mean().reset_index()

axs[1, 1].bar(result['LoanPurpose'], result['Default'], color='skyblue')
axs[1, 1].set_xlabel('Loan Purpose')
axs[1, 1].set_ylabel('Default Rate')
axs[1, 1].set_title('LoanPurpose - Default rate')
axs[1, 1].tick_params(axis='x', rotation=45)

# axs[1, 2].bar(...)
result = loan.groupby('HasCoSigner')['Default'].mean().reset_index()

axs[1, 2].bar(result['HasCoSigner'], result['Default'], color='skyblue')
axs[1, 2].set_xlabel('HasCoSigner')
axs[1, 2].set_ylabel('Default Rate')
axs[1, 2].set_title('HasCoSigner - Default rate')
axs[1, 2].tick_params(axis='x', rotation=45)

# Í∑∏ÎûòÌîÑ Í∞ÑÍ≤© Ï°∞Ï†ï
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Í∑∏ÎûòÌîÑ ÌëúÏãú
plt.show()
```

```python
loan = pd.DataFrame(loan)

# ÏÑúÎ∏åÌîåÎ°Ø Í∑∏Î¶¨Îìú ÏÉùÏÑ±
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
fig.suptitle('Default Rate by Numeric Features', fontsize=16)

# Í∞Å ÌäπÏÑ±Ïóê ÎåÄÌïú DefaultÏùò ÎπÑÏú®ÏùÑ histplotÏúºÎ°ú Í∑∏Î¶ΩÎãàÎã§.
numeric_features = ['Age', 'Income', 'LoanAmount', 'CreditScore', 'MonthsEmployed', 'NumCreditLines', 'LoanTerm', 'Education','InterestRate', 'DTIRatio']

for i, feature in enumerate(numeric_features):
    row, col = divmod(i, 5)
    sns.histplot(data=loan, x=feature, hue='Default', ax=axes[row, col], multiple='stack', bins=20, stat='percent')
    axes[row, col].set_title(feature)

# Í∑∏ÎûòÌîÑ Í∞ÑÍ≤© Ï°∞Ï†ï
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Í∑∏ÎûòÌîÑ ÌëúÏãú
plt.show()
```

## 9/30
### Í≥†ÏÜåÎìùÏ∏µÍ≥º Ï†ÄÏÜåÎìùÏ∏µ Î≤îÏ£ºÌôî
- Í≥†ÏÜåÎìùÏ∏µ: ÏÉÅÏúÑ 10%
- Ï†ÄÏÜåÎìùÏ∏µ: ÌïòÏúÑ 30% (ÏÉùÍ≥Ñ Í∏âÏó¨ ÏßÄÍ∏â Í∏∞Ï§Ä)

```python
# 'Income' ÏÉÅÏúÑ 10% ÌïÑÌÑ∞ÎßÅ
top_10_percent_income = loan[loan['Income'] >= loan['Income'].quantile(0.9)]

# 'Income' ÌïòÏúÑ 30% ÌïÑÌÑ∞ÎßÅ
bottom_30_percent_income = loan[loan['Income'] <= loan['Income'].quantile(0.3)]

# ÌååÏù¥Ï∞®Ìä∏Î•º Í∑∏Î¶¥ Ìï®Ïàò Ï†ïÏùò
def draw_pie_chart(data, ax, title):
    loan_purpose_counts = data['LoanPurpose'].value_counts()
    ax.pie(loan_purpose_counts, labels=loan_purpose_counts.index, autopct='%1.1f%%', startangle=140)
    ax.set_title(title)
    ax.axis('equal')  # ÌååÏù¥Ï∞®Ìä∏Î•º ÏõêÌòïÏúºÎ°ú Ïú†ÏßÄ

# 1x2 Í∑∏Î¶¨Îìú ÏÉùÏÑ±
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# ÏÉÅÏúÑ 10% IncomeÏùò ÌååÏù¥Ï∞®Ìä∏ Í∑∏Î¶¨Í∏∞
draw_pie_chart(top_10_percent_income, axs[0], 'Loan Purpose for Top 10% Income Earners')

# ÌïòÏúÑ 30% IncomeÏùò ÌååÏù¥Ï∞®Ìä∏ Í∑∏Î¶¨Í∏∞
draw_pie_chart(bottom_30_percent_income, axs[1], 'Loan Purpose for Bottom 30% Income Earners')

# Í∑∏ÎûòÌîÑ Í∞ÑÍ≤© Ï°∞Ï†ï
plt.tight_layout()

# Í∑∏ÎûòÌîÑ ÌëúÏãú
plt.show()
```

-> Ïú†ÏùòÎØ∏Ìïú Í≤∞Í≥ºÍ∞Ä ÎèÑÏ∂úÎêòÏßÄ ÏïäÏùå

### ÎÇòÏù¥Î≥Ñ LoanPurpose

```python
# 'Age'Î•º ÏÑ∏ Íµ¨Í∞ÑÏúºÎ°ú ÎÇòÎàÑÍ∏∞
loan['AgeGroup'] = pd.cut(loan['Age'], bins=[0, 30, 50, loan['Age'].max()], labels=['Young', 'Middle-aged', 'Senior'])

# ÌååÏù¥Ï∞®Ìä∏Î•º Í∑∏Î¶¨Îäî Ìï®Ïàò Ï†ïÏùò
def draw_pie_chart(data, ax, title):
    loan_purpose_counts = data['LoanPurpose'].value_counts()
    ax.pie(loan_purpose_counts, labels=loan_purpose_counts.index, autopct='%1.1f%%', startangle=140)
    ax.set_title(title)
    ax.axis('equal')  # ÌååÏù¥Ï∞®Ìä∏Î•º ÏõêÌòïÏúºÎ°ú Ïú†ÏßÄ

# 1x3 Í∑∏Î¶¨Îìú ÏÉùÏÑ±
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Í∞Å Íµ¨Í∞ÑÏóê ÎåÄÌïú ÌååÏù¥Ï∞®Ìä∏ Í∑∏Î¶¨Í∏∞
for i, age_group in enumerate(loan['AgeGroup'].unique()):
    data_subset = loan[loan['AgeGroup'] == age_group]
    draw_pie_chart(data_subset, axs[i], f'Loan Purpose for {age_group}')

# Í∑∏ÎûòÌîÑ Í∞ÑÍ≤© Ï°∞Ï†ï
plt.tight_layout()

# Í∑∏ÎûòÌîÑ ÌëúÏãú
plt.show()
```

# risk-taking
-> Ïà´ÏûêÍ∞Ä Ïª§ÏßàÏàòÎ°ù risk-takingÏùÑ Ìï† Í∞ÄÎä•ÏÑ±Ïù¥ ÎÜíÏùå
- 'MaritalStatus' == 1
- 'HasMortgage' == 1
- 'HasDependents' == 1
- 'HasCoSigner' == 1
Ïùº Îïå risk-taking Ìï† Í∞ÄÎä•ÏÑ±Ïù¥ ÎÇÆÏùå

```python
# MarriedÏùº Îïå 1, ÏïÑÎãê Îïå 0ÏúºÎ°ú Î∞îÍøà
loan['MaritalStatus'] = (loan['MaritalStatus'] == 'Married').astype(int)

loan.head(100)
```

## risk-taking column ÎßåÎì§Í∏∞: Îã§ ÎçîÌï¥ÏÑú ÏùåÏàòÎ°ú ÎßåÎì§Í≥† ÎçîÌïòÍ∏∞ 4 ÎÇòÎàÑÍ∏∞ 4

```python
loan['RiskTaking'] = (- (loan['HasMortgage'] + loan['MaritalStatus'] + loan['HasDependents'] + loan['HasCoSigner']) + 4) / 4

loan.head(100)
```

```python
# 'RiskTaking'Ïóê ÎåÄÌïú 'Default' ÎπÑÏú® Í≥ÑÏÇ∞
default_rate_by_risktaking = loan.groupby('RiskTaking')['Default'].mean().reset_index()

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(10, 6))
sns.barplot(x='RiskTaking', y='Default', data=default_rate_by_risktaking, palette='viridis')
plt.xlabel('RiskTaking')
plt.ylabel('Default Rate')
plt.title('Default Rate by RiskTaking')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Í≤ΩÏ†úÏ†Å ÏïàÏ†ïÏÑ± (Economic Stability)
- Í≤∞ÌòºÌïú ÏÉÅÌÉú (Î≤îÏ£º)
- ÏÜåÎìùÏù¥ ÎÜíÏùå (Ïó∞ÏÜç)
- full-timeÏùº Îïå (Î≤îÏ£º)
- month employed (Ïó∞ÏÜç)

```python
# full-timeÏùº Îïå 1, ÏïÑÎãê Îïå 0ÏúºÎ°ú Î∞îÍøà
loan['EmploymentType'] = (loan['EmploymentType'] == 'Full-time').astype(int)

loan.head()
```

```python
loan['EmploymentTypeMaritalStatus'] = loan['EmploymentType'] + loan['MaritalStatus'] + 1

loan.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# 'Income'Í≥º 'MonthsEmployed'Î•º 1Í≥º 100 ÏÇ¨Ïù¥Î°ú Ïä§ÏºÄÏùºÎßÅ
scaler = MinMaxScaler(feature_range=(1, 100))
scaled_columns = ['Income', 'MonthsEmployed']
loan[scaled_columns] = scaler.fit_transform(loan[scaled_columns])

# 'Income', 'MonthsEmployed', 'EmploymentTypeMaritalStatus' Ïª¨ÎüºÏùÑ Í≥±ÌïòÍ≥† ÏÑ∏Ï†úÍ≥±Í∑º Ï†ÅÏö©ÌïòÏó¨ 'EconomicStability' Ïó¥ ÏÉùÏÑ±
loan['EconomicStability'] = np.cbrt(loan['Income'] * loan['MonthsEmployed'] * loan['EmploymentTypeMaritalStatus'])

loan.head()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=loan, x='EconomicStability', hue='Default', bins=10, kde=True, palette='Set1')
plt.xlabel('EconomicStability')
plt.ylabel('Count')
plt.title('Distribution of Default by EconomicStability')
plt.legend(title='Default', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
correlations = loan.corr()['Default']

correlations
```
