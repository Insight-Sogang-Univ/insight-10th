# 1. 통계란?
데이터 수집/분석/해석 과정 중 패턴/관계/변동 이해해 결론을 도출하는 학문 및 방법론
## 1) 기술 통계학
- 기술통계학 <br>
데이터 요약 / 기술 > 평균/중앙값/표준편차/분포<br>
    - 수치기술 통계 : 중심 위치 척도 / 변동성 척도 / 연관성 척도 
    - 표와 그래프 기술 통계 : EDA/시각화 


## 2) 추론 통계학
현재 데이터 기반 미래 정보 추론 및 결론 도출 기법 <br>
표본을 기반으로 모집단 정보 추측 > 통계적 가설검정, 신뢰구간 

# 2. 수치 기술 통계
## 1) 중심 위치 척도
- 평균 mean<br>
데이터 값 총합 / 데이터 개수 <br>
데이터 중심 경향성 보이며, 이상치에 민감<br><br>
- 중앙값 median <br>
데이터 크기 순으로 정렬 후 가운데 위치한 값 <br>
데이터 중심 경향성 보이며, 이상치에 덜 민감<br><br>
- 최빈값 mode<br>
데이터에서 가장 자주 나타나는 값<br>
주로 범주형 데이터에서 이용, 분포 이해에 도움 <br>
> 이를 통해 **자료 개형 확인** 가능 : 왜도 Skewness<br>

- Positive Skewness<br>
오른쪽 꼬리가 왼쪽보다 길 때<br>
평균, 중앙값 > 최빈값<br><br>
- Negative Skewness<br>
왼쪽 꼬리가 오른쪽보다 길 때 <br>
평균, 중앙값 < 최빈값<br><br>
- Normal distribution<br>
오른쪽, 왼쪽 대칭<br>
평균,중앙값 = 최빈값<br><br>
## 2) 변동성 척도 
- 사분위수 quartile <br>
데이터의 표본을 4개의 동읠 부분으로 나눈 값 <br>
    - Q1 = 1+(n-1) * 0.25
    - Q2 = 중앙값
    - Q3 = 1+(n-1) * 0.75
<br><br>
- 사분위간 범위 IQR <br>
Q3에서 Q1값을 제외한 범위<br>
박스 플롯에서 박스로 표현되는 부분!
<br>
- 범위 range <br>
데이터 최대값과 최솟값의 차이 나타냄 > 데이터 전체 변동 범위 알려줌
<br><br>
- 분산 variance <br>
평균으로부터 데이터 값이 얼마나 분포되어 있는지 보여주는 측정 지표<br>
(데이터 값 - 평균)** 2의 합<br><br>


- 표준편차 standard deviation<br>
평균으로부터 데이터 값이 얼마나 분포되어 있는지 나타내는 통계적 지표<br>
분산의 제곱근 <br><br>

- 변동계수 coefficient of variation <br>
데이터의 상대적 변동성 나타내는 통계적 지표<br>
변동계수 % = 표준편차 / 평균 * 100<br>


## 3) 연관성 척도 
- 공분산 covariance
두 변수가 각 평균으로부터 떨어진 값 서로 곱한 후 평균 값으로 계산<br>
어떤 방향으로 변화하는지와 그 변화의 크기 표현 <br><br>
Cov(x,y) > 0 : X-Y 양의 관계<br>
Cov(x,y) < 0 : X-Y 음의 관계<br>
Cov(x,y) = 0 : X-Y 양도 음도 아닌 관계<br>
=> 이 경우 서로 관계가 있는 경우도 존재<br>
<br>
- 상관계수 correlation 
두 변수 간 선형적 관계 나타내는 통계적 지표 <br>
-1 < 상관계수 < 1<br>
상관관계와 인과관계는 다르다!<br>
공분산을 -1,1사이로 표준화 한 것 <br><br>
Corr(x,y) = 1 에 가까울수록 양의 상관관계, 1이면 완전한 양의 상관관계<br>
Corr(x,y) = -1 에 가까울수록 음의 상관관계, -1이면 완전한 음의 상관관계<br>
<br>


# 3. 추론 통계 - 모집단 
전체 집단을 조사하는 것이 현실적으로 불가능한 경우 모집단에서 일부 사람들을 추출해 표본 형성 후 표본 집단을 바탕으로 통계량 계산 <br>
-> 모집단과 모수 정확히 알 수 없지만 표본과 표본 통계량 통해 **추정**가능 
<br>
<br>
- 대수의 법칙
표본 크기 커질수록 표본 평균이 모집단에 가까워짐 <br>
취합 표본 수 많을수록 통계 정확도 올라감 <br>
=> 데이터 많을수록 좋다!<br>
- 중심극한 정리
표본 크기 크면 표본 평균의 분포가 정규분포에 가까워짐<br>
=> 모집단 분포와 상관 X<br>

## 1) 확률 분포
확률 변수가 특정 값 가질 확률을 나타내는 함수 <br>
대표적으로 **이산확률분포**와 **연속확률분포**<br><br>
- 이산확률분포
확률 변수가 이산적 값 가지는 경우<br>
유한하거나 셀 수 있는 값을 가짐<br>
<br>
- 연속확률분포
확률 변수가 연속적인 값을 갖는 경우<br>
무한히 많은 값의 범위 가짐, 그 값 중 하나를 선택하는 건 불가능에 가까움<br>

<br>

## 2) 예시
### [1] 정규분포
종 모양, 좌우대칭<br>
평균에 의해 위치정해짐, 표준편차의 크기에 의해 모양 결정<br><br>


### [2] T분포
표본 평균이 어떤 분포를 가지는지 보임<br>
표준정규분포처럼 0을 중심으로 종모양, 대칭 분포<br>
꼬리는 표준정규분포보다 두꺼움<br>
자유도 n에 따라 모양 변함, n이 커질수록 표준정규분포N(0,1)에 수렴<br><br>


### [3] 카이제곱분포
범주형 데이터 분석/카이제곱 검정 등 **통계적** 검정에서 활용 <br>
정규분포 제곱 = 카이스퀘어분포 > 제곱으로 **양수** 값 가짐<br>
자유도 따라 분포 모양 변화, 자유도 커질수록 평균/분산 증가<br><br>


### [4] F분포
두 개의 카이제곱분포 비율 관련 확률분포 <br>
분산/회귀분석에서 사용, 두 모집단의 분산의 동일성 여부 비교 등 **통계적** 검정에서 활용 <br>
자유도 증가할수록 F분포 확률밀도 함수 모양이 정규분포 모양으로, 분산 감소<br><br>


### [5] 이항분포 + 베르누이 분포 
연속된 n번의 독립시행에서 각 시행이 성공할 확률 p를 가질 때 만들어지는  **이산확률분포**<br>
ex. 동전던지기 <br>
이항분포 시행이 1번 = **베르누이분포**<br>
=> 시행 횟수 증가할수록 분포 대칭적으로<br><br>


### [6] 포아송 분포
단위 시간 안에 어떤 사건이 몇 번 일어날 지 나타내는 **이산확률분포**<br>
-> 시행 횟수 증가할 수록 분포 대칭적으로<br><br>


# 4. 가설 검정 
## 1) 가설 검정 절차
**가설검정 > 유의 수준 설정 > 검정통계량 산출 > 가설 기각/채택 판단 **<br>
- 가설 설정
귀무가설 H0 - 대립가설 H1 설정<br>
귀무가설 = 확인하기 용이 or 기각하고자 하는 명제로 <br>
<br>
- 유의 수준 설정 
귀무가설 오류 판단 기준 = 유의 수준 <br>
 α가 유의 수준, 1-α가 신뢰 수준 <br>
 α값 커질수록 채택 가능성 높아지며, 반대로 작아질수록 대립가설 기각 가능성 증가
<br><br>
- 검정통계량 산출
가설 검정 결과 판단 시 사용<br>
데이터 분포 및 특징에 따라 Z통계량, T통계량 중 선택
<br><br>
- 가설 기각/채택 판단 
검정통계량이 신뢰구간에 위치 : 귀무가설 채택 & 대립가설 기각 <br>
    벗어날 경우 귀무가설 기각 & 대립가설 채택 <br>
유의 확률 p-value : 검정통계량에서 유도 가능 <br>
유의 확률이 유의 수준보다 클 경우 귀무가설 채택<br>
    벗어날 경우 대립가설 채택 


## 2) 1종, 2종 오류
- 1종 오류 (𝛼) : 귀무 가설이 실제 참이었는데 기각할 확률<br>
- 2종 오류 (𝛽) : 귀무 가설이 실제 거짓이었는데 채택할 확률<br>

# 5. 조건부 확률과 베이즈 정리 
## 1) 조건부 확률
어떤 사건이 일어났을 때 연관 사건이 일어날 확률 <br>
**P(B|A)** : A가 일어났을 때 B도 일어날 확률

## 2) 베이즈 정리 
조건부 확률에 포함된 정보를 이용해 새로운 정보 추론 <br>

사후확률 = (우도 * 사전확률) / 증거
