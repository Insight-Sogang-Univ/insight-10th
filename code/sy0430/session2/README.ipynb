# 2. Feature 분석
## 1) Feature 종류
- 범주형
    - 명목형 : 순서 X, 단순 분류
    - 순서형 : 범주형 중 순서 관계 O
<br>
    
- 수치형(이산형, 연속형)
    - 이산형 : 이산적 값, 정수 단위, 셀 수 있음
    - 연속형 : 연속 값 갖는 데이터 



## 2) Feature 유형에 따른 시각화 방법
- 시각화 전 데이터 특성 파악 필요 <br>
**범주형** vs **수치형**<br>
결측값 유무 <br>
<br>
- 데이터 특성 파악 후 <br>
**명목형** :Bar plot / Count plot / Pie chart<br>
**자료 개형** : Histogram / Box plot / Stem and leaf plot<br>
**시계열 / 변수관계** : Time series plot / Scatter plot<br>

## 3) 사용 라이브러리

Numpy 라이브러리 활용 플로팅 라이브러리<br>
<br>
- Seaborn<br>
matplotlib 기반 통계데이터 시각화 최적화 라이브러리<br>
<br>

- 히스토그램을 곡선으로 연결한 그래프

## 3) 막대그래프 barplot / 포인트플롯 pointplot
- 막대그래프 : 범주형 데이터 값에 따라 수치형 데이터 값이 달라지는 양상 파악<br>
x='범주형 데이터', y='수치형 데이터'<br>
<br>
- 포인트플롯 : 막대 그래프와 모양만 다름, 동일 정보


## 4) 박스플롯 boxplot
- 데이터 분포, 중앙값, 이상치 시각적으로 표현<br>
- 최댓값 / 제 3사분위수 / 제 2사분위수 / 제 1사분위수 / 최솟값 제공<br>
- 이상점 제공

## 5) 바이올린 플롯 violinplot
- 박스플롯 + 커널밀도추정함수 그래프<br>
- 박스플롯보다 정확한 분포 파악 가능 <br>
sns.violinplot(x='칼럼명', y='칼럼명',[hue='범주'] ,data=data, [split = True])

## 6) 카운트 플롯 countplot
- 카테고리별 데이터 개수 수치 표시<br>
sns.countplot(x='칼럼명',data=data) => y값은 수치<br>
sns.countplot(y='칼럼명',data=data) => x값은 수치 


## 7) 파이 그래프 pie
- 부분:전체 간 비율 표현 = **비율 강조**<br>
plt.pie(x=[비율 값],labels=[라벨 명], autopct='%.1f%%')


## 8) 라인 그래프 line plot
