# 주관식 문제

## 명시적 데이터와 암묵적 데이터의 장단점을 2개씩 쓰시오. (장점 2개, 단점 2개)

answer  
명시적 장점: 당시의 사용자의 취향이 정확하게 반영됨
<br>명시적 단점: 사람의 취향은 변하기 마련임. 지금은 바뀌었을 수 있다.
<br>암묵적 장점: 데이터를 얻기 쉬움
<br>암묵적 단점: 사람의 취향을 직접적으로 알 수 없다.

## 협업 필터링에 속하는 필터링 2가지를 쓰시오. (큰 분류 2가지)

answer: 사용자 기반 / 아이템 기반
